```python
import requests as r
import pandas as pd
import time
```

```python
uri = 'http://localhost:8088/toid/osgb100000456'
```

```python
res_arr = []
init = time.perf_counter()
for i in range(1, 5000):
    ids = 5000 + i
    res = r.get(f'{uri}{ids}')
    res_arr.append(res.json())

print(f'total exec time: {time.perf_counter() - init} seconds')
```

```python
len(res_arr)
```

```python
df = pd.io.json.json_normalize(res_arr)
```
